```python
import matplotlib.pyplot as plt
x=[12, 17, 32, 39]
motivo=['Produto de outra empresa era melhor', 'Produto mais barato', 
        'No receberam ateno adequada', 'Atendimento rude']
cores=['y', 'g', 'b', 'r']
plt.axis('equal')
plt.pie(x,labels=motivo, colors=cores, shadow=True,
        autopct='%1.1f%%')
plt.title('Qualidade do Produtos e Servios')
plt.show()
```

```python
import matplotlib.pyplot as plt
x=['Produto de outra empresa era melhor', 'Produto mais barato', 
        'No receberam ateno adequada', 'Atendimento rude']
y=[12, 17, 32, 39]
plt.bar(x,y)
plt.xticks(rotation=45, ha='right')
plt.title('Qualidade do Produtos e Servios')
plt.show()
```
